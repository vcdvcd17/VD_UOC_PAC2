```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("credit_risk_dataset.csv")
```

```python
df["LTI_bin"] = pd.cut(df.loan_percent_income,8)
```

```python
pt = df.pivot_table(index=["loan_intent"], columns=["LTI_bin"], values=["loan_status"], aggfunc=np.mean)
```

```python
pt.columns = df["LTI_bin"].cat.categories
pt.columns.name = "Loan to Income"
pt.index.name = "Loan Intent"
```

```python
pt.columns
```

```python
plt.figure()
plt.title("Loan default rate")
sns.heatmap(pt)
plt.savefig("loan_defaultrate_heatmap.png", bbox_inches = "tight");
```
